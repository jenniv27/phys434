```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (15,10)
```

```python
d = stats.norm.rvs(loc = 5., scale = 0.01, size = 100000)
#norm = normal distribution function
#rvs = random variates
#loc = 5 is the mean
#scale = 0.01 is the standard deviation (in stats.norm)
#size = 100000 is the number of data points
```

```python
fig, ax = plt.subplots(1, 1)
#creates figure (fig) and set of subplots (ax) objects
ax.hist(d,50, density = True)
#take the subplot ax and create a histogram
    #using the 100000 data points (d) 
    #with 50 bins
    #density = True normalizes the bin heights so that the 
    #integral of the histogram is 1
plt.tick_params(labelsize = 24)
#labelsize = float or string; set labels of each axis to be size 24 font
plt.xlim([4.95,5.05])
#change limits of x axis
x = np.linspace(4.95,5.05,1000)
#generates 1000 evenly spaced numbers between 4.95 and 5.05
ax.plot(x,stats.norm.pdf(x,loc = 5., scale = 0.01), linewidth = 8, alpha = 0.7)
#take the subplot ax and add to it a plot (in orange line) such that:
    #x (defined previously, 1000 points) is given as x-values
    #y is the pdf with mean = 5 and stand.dev is 0.01
    #linewidth is self-explanatory
    #alpha is plot transparency
plt.show()
#finally, display the plot
```

```python
ax = plt.hist(d,50)
#generates a single plot (histogram)
    #using d as data, and 50 bins. 
    #by default, density = False so this plot is not normalized
    
#the following code directly manipulates the global parameters of matplotlib
    #pyplot (plt):
plt.yscale('log')
#change the y axis to a log scale
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
#again working with the objects fig and ax
ax.hist(d,50, density=True)
plt.yscale('log')
plt.tick_params(labelsize = 24)
plt.xlim([4.95,5.05])
x = np.linspace(4.95,5.05,1000)
ax.plot(x,stats.norm.pdf(x,loc = 5., scale = 0.01),linewidth = 8,alpha = 0.7)
plt.show()
#recreates the first graph but now the y axis is a log scale
```
